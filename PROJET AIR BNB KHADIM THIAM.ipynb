```python
%matplotlib inline 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
missing_values = ["n/a", "na", "--"]
data = pd.read_csv('listings.csv', na_values=missing_values)
#cal = pd.read_csv('calendar.csv')
```

```python
arr = data[['id','neighbourhood']].groupby('neighbourhood').mean().round().sort_values(by='id', ascending=False)
arr.reset_index(0, inplace=True)
arr.tail()
```

```python
plt.figure(figsize=(18,15))
sns.barplot(x=arr['id'], y=arr['neighbourhood'], palette="Blues_r")
plt.xlabel('listings_id ', fontsize=15, color='#2980b9')
plt.ylabel('neighbourhood ', fontsize=15, color='#2980b9')
plt.title("listings by neighbourhood", fontsize=18, color='#3742fa')
plt.xticks(rotation= 45)
plt.tight_layout()
```

```python
arr = data[['id','property_type']].groupby('property_type').mean().round().sort_values(by='id', ascending=False)
arr.reset_index(0, inplace=True)
arr.head()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=arr['id'], y=arr['property_type'], palette="Blues_r")
plt.xlabel('listings_id ', fontsize=15, color='#2980b9')
plt.ylabel('property_type ', fontsize=15, color='#2980b9')
plt.title("listings by property_type", fontsize=18, color='#3742fa')
plt.xticks(rotation= 45)
plt.tight_layout()
```

```python
from folium.plugins import marker_cluster
def generateBaseMap(default_location=[48.85100,2.35869], default_zoom_start=10):
    base_map = folium.Map(location=default_location, control_scale=True, zoom_start=default_zoom_start)
    return base_map
```

```python
base_map = generateBaseMap()
marker_cluster.MarkerCluster(data=data[['latitude', 'longitude']].values.tolist(), radius=8, max_zoom=13).add_to(base_map)
```
